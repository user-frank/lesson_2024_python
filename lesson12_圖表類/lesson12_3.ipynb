```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
axes = fig.add_subplot(1, 1, 1)
subjects = ['chinese','english','math','history','discover']
nums1 = [1, 4, 9, 16, 25]
nums2 = [25, 16, 12, 4, 1]
axes.plot(subjects,nums1,'+--y',label='studen1')
axes.plot(subjects,nums2,'o-r',label='student2')
axes.set_title('students')
axes.set_xlabel('subject')
axes.set_ylabel('scores')
axes.grid(axis='y')
axes.set_xticks([0,1,2,3,4])
axes.set_xticklabels(['ch','en','ma','hi','cover'])
axes.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
  #外框標示
def Appearance():  
    chart.set_title('department_performance')
    chart.set_xlabel('Days')
    chart.set_ylabel('performance')
    chart.grid()
    chart.set_xticks(days)

#設置邊框
def charts():
    chart.spines['top'].set_edgecolor('red')    # 上邊框設為紅色
    chart.spines['top'].set_linewidth(5)  #這裡是邊框粗細

    chart.spines['right'].set_edgecolor('blue')  # 右邊框設為藍色
    chart.spines['right'].set_linewidth(2)      

    chart.spines['bottom'].set_edgecolor('green')  # 下邊框設為綠色
    chart.spines['bottom'].set_linewidth(10)        

    chart.spines['left'].set_edgecolor('orange')   # 左邊框設為橙色
    chart.spines['left'].set_linewidth(2) 

#設置線的顏色
def line_color():
    chart.plot(days[:15], celsius[:15], 'o--r', label='A_Team')  
    chart.plot(days[15:], celsius[15:], 's--b', label='B_Team')
    chart.legend(loc='upper left')   

# 主程式
if __name__ == "__main__":
    fig = plt.figure(figsize=(10, 6))
    chart = fig.add_subplot(1,1,1)

    celsius = [round(random.uniform(28.5, 35.2),1) for _ in range(1,31)]
    days = [num for num in range(1,31)]
    #調用所有def  
    Appearance()
    charts()
    line_color()
    #運行
    plt.show()

```
